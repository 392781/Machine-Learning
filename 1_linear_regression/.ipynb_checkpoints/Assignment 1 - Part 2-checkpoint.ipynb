```python
from __future__ import print_function
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn import linear_model
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reads in CSV file
data = pd.read_csv("Video_Games_Sales.csv")

# Takes only NA sales, Global Sales, and Critic + User score columns
review_data = data[["EU_Sales", "Global_Sales", "Critic_Score", "User_Score"]]
# Drops rows with NaN values
review_data = review_data.dropna()
# Drops rows with "tbd" string (Was an issue)
review_data = review_data[review_data.User_Score.str.contains("tbd") == False]

# X attributes are everything but global sales
X = review_data.drop("Global_Sales", axis = 1)
# Target value is the global  sales
Y = review_data["Global_Sales"]

# Training/Testing split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 100)
```

```python
'''
    Create a LinearRegression object and use it to find fit using training sets
    Then use testing data set to find the prediction target set
'''
regress = linear_model.LinearRegression()
regress.fit(X_train, Y_train)
Y_pred = regress.predict(X_test)

'''
    Print out of intercept + coefficients
'''
print("Intercept:\n", regress.intercept_)
print("\nCoefficients:\n", regress.coef_)

'''
    Calculation of training error using N elements of the training set
    Calculation of testing error using target test and target prediction
'''
train_error = mean_squared_error(Y_train[0:Y_pred.shape[0]], Y_pred) / 2
test_error  = mean_squared_error(Y_test, Y_pred) / 2

'''
    Print out of error
'''
print("\nTrain Error:\n", train_error)
print("\nTest Error:\n", test_error)

'''
    Focused plot of the first 10% of the testing target and prediction target
    to show a clearer view of the prediction data
'''
plt.style.use("ggplot")
plt.figure(figsize=(15,7))
plt.plot(range(0, Y_test.shape[0]//10), Y_test[0:Y_test.shape[0]//10], linewidth = 1, alpha = 0.7)
plt.plot(range(0, Y_test.shape[0]//10), Y_pred[0:Y_test.shape[0]//10], linewidth = 1, alpha = 0.5)
plt.legend(["Test", "Prediction"])
plt.xlabel("Data")
plt.ylabel("Sales (Millions)")

plt.show()

'''
    Plot of the entire testing and prediction target sets
'''
plt.style.use("ggplot")
plt.figure(figsize=(15,7))
plt.plot(range(0, Y_test.shape[0]), Y_test, linewidth = 1, alpha = 0.7)
plt.plot(range(0, Y_test.shape[0]), Y_pred, linewidth = 1, alpha = 0.5)
plt.legend(["Test", "Prediction"])
plt.xlabel("Data")
plt.ylabel("Sales (Millions)")

plt.show()
```
